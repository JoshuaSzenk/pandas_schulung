```python
import pandas as pd

pd.set_option("display.max_rows", 5)
reviews = pd.read_csv("winemag-data-130k-v2.csv", index_col=0)

print("Setup complete.")

reviews.head()
```

```python
median_points = reviews.points.median()

median_points
```

```python
countries = reviews.country.unique()

countries
```

```python
reviews_perCountry = reviews.country.value_counts()

reviews_perCountry
```

```python
centered_price = reviews.price - reviews.price.mean()

centered_price
```

```python
bargain_wine = reviews.loc[(reviews.points / reviews.price).idxmax(), "title"]

bargain_wine
```

```python
n_trop = reviews.description.map(
    lambda desc: "tropical" in desc
).sum()  # Summe tropical in description
n_fruity = reviews.description.map(
    lambda desc: "fruity" in desc
).sum()  # Summe fruity in description

description_counts = pd.Series([n_trop, n_fruity], index=["tropical", "fruity"])

description_counts
```

```python
def berechne_sterne(row):

    if row.country == "Canada":
        return 3
    elif row.points >= 95:
        return 3
    elif row.points >= 85:
        return 2
    else:
        return 1
```

```python
star_ratings = reviews.apply(berechne_sterne, axis="columns")

star_ratings
```

```python
###Musterlösung Kaggl Aufgabe 7

star_ratings = reviews.points.map(lambda p: 3 if p >= 95 else (2 if p >= 85 else 1))

# 2. Die Kanada-Sonderregel nachträglich drüberbügeln
star_ratings[reviews.country == "Canada"] = 3

star_ratings
```
